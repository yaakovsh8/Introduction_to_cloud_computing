```python
!pip install firebase

import json
import ipywidgets as widgets
from IPython.display import display
from firebase import firebase

from google.colab import drive
drive.mount('/content/drive')

# Path to the JSON file in your Google Drive
json_file_path = '/content/drive/My Drive/students.json'

# Read the JSON file
with open(json_file_path, 'r') as file:
    data = json.load(file)

# Initialize Firebase connection
FBconn = firebase.FirebaseApplication('https://salamandra-714f6-default-rtdb.firebaseio.com/', None)

# Function to upload data from JSON to Firebase
def upload_data(data_list): # Changed the parameter name to reflect that it's a list
    for data in data_list: # Iterate over each dictionary in the list
        # Prepare data to upload
        data_to_upload = {
            'TV-Show': data.get('TV_Show', ''),
            "link": data.get('link', ''),
            "Courses": data.get('Courses', []),
            "Email": data.get('Email', ''),
            "last name": data.get('last_name', ''),
            "Name": data.get('Name', '')
        }

        # Upload data to Firebase
        result = FBconn.post('/ex5_post/', data_to_upload)
        print(result)

# Call the function to upload data, passing the list of dictionaries
upload_data(data)
```
